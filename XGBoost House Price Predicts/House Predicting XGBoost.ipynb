BUILDING XGBOOST MODEL 


```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='price')
y = df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=100)
```

SCALLING NUMERICAL DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

BUILDING XGBOOST DMATRIX MODEL

```python
import xgboost as xgb
```

```python
model = xgb.DMatrix(X_train, y_train)
params = {'colsample_bytree':0.7, 'max_depth':3, 'eta': 0.1, 'subsample': 0.8, 'objective': 'reg:linear', 
            'min_child_weight': 1}
model = xgb.train(params, model)
```

```python
dmat_test = xgb.DMatrix(X_test)
y_preds = model.predict(dmat_test)
y_preds
```

Accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error
import math 
result_dmat = mean_absolute_percentage_error(y_test.values, y_preds)
result_dmat
```

BUILDING XGBOOST REGRESSOR

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor(max_depth = 15, learning_rate = 0.03, 
                random_state = 20, num_parallel_tree = 4, base_score = 0.2)
model_xgb.fit(X_train, y_train, early_stopping_rounds=20, eval_set=[(X_test, y_test)], verbose=False)
```

RESULT ACCURACY

```python
from sklearn.metrics import mean_absolute_percentage_error

predict = model_xgb.predict(X_test)
result = mean_absolute_percentage_error(y_test, predict)
print('Mean absolute percentage Error Rate:', result*100,'%')
```
